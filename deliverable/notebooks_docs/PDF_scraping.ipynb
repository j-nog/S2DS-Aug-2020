## Extracting the coordinates and repositioning

Here we identify and reconstruct any mismatched boxes which might have cut lines back into the same line by sorting the "line boxes" (see example figure below). This usually happends when multiple columns or indents of text are involved and allways happends with tables. $$\hskip{1.5in} (x_1,y_1)\hskip{1.5in} (x_1,y_1)\\\fbox{$\fbox{Box 1, linebox 1}\\\fbox{box 1 linebox 2}$}\fbox{$\fbox{box 2 linebox 1. }\\\fbox{box 2 linebox 2}$}\\(x_0,y_0)\hskip{1.5in}(x_0,y_0)\hskip{1.5in}$$

**Figure: Here $(x_0,y_0)$ and $(x_1,y_1)$ indicate the box positions of box 1 and box 2, written in the table as box_pos $(x_0,y_0,x_1,y_1)$ respectively. There are also boxes around each of the lines contained in each box, these "lineboxes" also have coordinates which we stored as line_pos. Here we see a line containing the text "Box 1, linebox 1 box 2 linebox 1." being cut into two boxes, this occurs often when reading tables but also due to erronous reading of indentations.**

These boxes have coordinates measured in points starting the measurement from the bottom left corner of the page. Since points are genarally not of well difined hight (https://en.wikipedia.org/wiki/Point_(typography)) we just use the page size reference A4 which has 595 × 842 points as a ballpark estemate. It is easily confirmed that the coordinates fall into this measurment system.

## Detect footers, headers & margin notes

Assuming that the footer is going to be in the same position in the vertical and hoizontal alignment (i.e. no ireggular patterns such as LaTeX footnotes occur) sort the lines by bottom left coordinate and (rounded down). Take out the bottom line and calculate the overlap of the strings entering this table using the difflib ratio method.

````python
no_breaks = df_form[df_form['text_type'].notnull()].round({'y0_ln':1,'x0_ln':1,'y1_ln':1,'x1_ln':1})

# Kills bottom line by sorting and aligning by bottom left coordinate of 
# sentence box. If lines are very similar (overlap) measured by tolerance
# drop the lines.

def check_empty_iterator(iterator):
    try:
        next(iterator)
    except StopIteration:
        pass

def kill_footers(tolerance = 0.9):
    bottom_ln = no_breaks.round({'y0_ln':1})['y0_ln'].min()
    to_match = no_breaks[no_breaks.y0_ln==bottom_ln].sort_values(by='x0_ln')['text']
    it1 = to_match.iteritems()
    it2 = to_match.iteritems()
    overlap = []
    check_empty_iterator(it2)
    for step in range(len(to_match)-1):
        value = next(it1)
        next_value = next(it2)
        match = difflib.SequenceMatcher(None,value[1],next_value[1]) 
        overlap.append(match.ratio())
    if overlap:
        if(sum(overlap)/len(overlap)) > tolerance:
            df_form.drop(to_match.index,inplace = True)
    else: 
        print('no footer with tolerance level {}'.format(tolerance))

        
def kill_headers(tolerance = 0.9):
    top_ln = no_breaks.round({'y1_ln':1})['y1_ln'].max()
    to_match = no_breaks[no_breaks.y1_ln==top_ln].sort_values(by='x1_ln')['text']
    it1 = to_match.iteritems()
    it2 = to_match.iteritems()
    overlap = []
    check_empty_iterator(it2)
    for step in range(len(to_match)-1):
        value = next(it1)
        next_value = next(it2)
        match = difflib.SequenceMatcher(None,value[1],next_value[1]) 
        overlap.append(match.ratio())
    if overlap:
        if(sum(overlap)/len(overlap)) > tolerance:
            df_form.drop(to_match.index,inplace = True)
    else: 
        print('no header with tolerance level {}'.format(tolerance))
````

## Indexing sections etc

Filling empty values with respective type for riffling

````python
df_form['text_type']=df_form['text_type'].fillna(method= 'bfill').fillna(method = 'ffill')
df_form
````

Identify if there is a change of text type

````python
def riffle_repeated_type():
    notnull_df = df_form[df_form['text_type'].notnull()]
    return (notnull_df[['text_type']] == notnull_df[['text_type']].shift()).any(axis=1)

text_type_change= ~riffle_repeated_type().reindex(df_form.index).fillna(True)
text_type_change[0]=False
df_form['type_change']=text_type_change
df_form
````

If there is no change of text type join sentences else use a counter to track in which (sub) section the text is located. 

````python
notnull_df = df_form[df_form['text_type'].notnull()]
u_index = df_form.sort_values(by = 'curr_linefontsize').groupby(notnull_df['curr_linefontsize'].round(decimals = 0))['text_type'].unique().drop(textsize)
u_index = [u_index.iloc[ntry][0] for ntry in range(u_index.size)]
sec_counters = []
[sec_counters.append(0) for section_level in range(len(u_index))]
# print(pd.DataFrame(np.transpose(np.array([u_index,sec_counters]))))
# counter_df =pd.DataFrame([u_index,sec_counters],index=['name','count']).T
# if counter_df[counter_df['name']=='suptext'].index.values==0: 
#     print('thing')
u_index
````

````python
def create_section_index():
    """
    Creates new columns automaticaly for levels of sub(secitons)
    in text.
    """
    notnull_df = df_form[df_form['text_type'].notnull()]
    u_index = df_form.sort_values(by = 'curr_linefontsize').groupby(notnull_df['curr_linefontsize'].round(decimals = 0))['text_type'].unique().drop(textsize)
    u_index = np.array([u_index.iloc[ntry][0] for ntry in range(u_index.size)])
    sec_counters = []
    [sec_counters.append(0) for section_level in range(len(u_index))]
    counter_df = pd.DataFrame([sec_counters],columns = u_index)
    # appending to df_form
    df_form[u_index] = counter_df #pd.concat([df_form,counter_df],axis=1)
    df_form[u_index] = df_form[u_index].ffill(0)
    return u_index

def iter_the_cols(df_in,list_ind):
    """
    Adds 1 to the column generatied by df.create_secton_index() 
    (say subsection) if a change of text type occurs (say from 
    susection to main_text). Total section index obtained by cumsum 
    over these columns
    """
    if df_in['type_change'] == True: 
        if df_in['text_type'] in df_in.keys():   
            level = np.where(list_ind==df_in['text_type'])[0]
            df_in[df_in['text_type']] =1
            for entry in range(level[0]):
                df_in[list_ind[entry]]=0
            #df_in[list_ind]=df_in[list_ind].ffill()
    return df_in

u_index = create_section_index()
df_form = df_form.apply(iter_the_cols,list_ind=u_index,axis=1)
df_form[u_index[-1]]=df_form[u_index[-1]].cumsum()
for depth in range(len(u_index)-1):
    df_form[u_index[-2-depth]]=df_form.groupby(u_index[-1-depth])[u_index[-2-depth]].cumsum()
````
