## Raw text for each sections in the PDF file

```python
def get_section_raw_text(titles, goto_pages):
    """
    Utility function to extract raw text of each section
    Return a list containing [section_title, section_raw_text]
    """
    section_raw_text = []
    for i, title in enumerate(titles):
        temp_txt = ""    
        if i == len(titles)-1:    
            txts = df_layout[df_layout.page_no >= goto_pages[i]]["text"]
            copy_txt = False
            for j in range(len(txts)):
                #if txts.iloc[j].strip() == titles[i].strip():
                if txts.iloc[j].startswith(titles[i]):
                    copy_txt = True
                    continue
                elif copy_txt:
                    temp_txt += txts.iloc[j]
            section_raw_text.append((titles[i].strip(), temp_txt))
            break
        else:
            txts = df_layout[(df_layout.page_no >= goto_pages[i]) & (df_layout.page_no <= goto_pages[i+1])]["text"]
            copy_txt = False
            for j in range(len(txts)): 
                #if txts.iloc[j].strip() == titles[i].strip():
                if txts.iloc[j].startswith(titles[i]):
                    copy_txt = True
                    continue
                #elif txts.iloc[j].strip() == titles[i+1].strip():
                elif txts.iloc[j].startswith(titles[i+1]):
                    copy_txt = False
                    continue
                elif copy_txt:
                    temp_txt += txts.iloc[j]
            section_raw_text.append((titles[i].strip(), temp_txt))
    return section_raw_text                
```

```python
titles = [cont for cont in df_content.content_title]
goto_pages = [page for page in df_content.goto_page]

section_raw_text = get_section_raw_text(titles, goto_pages)
df_section_raw_text = pd.DataFrame(section_raw_text, columns=["section_title", "raw_text"])
```
