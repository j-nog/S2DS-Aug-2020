```python
def find_keywords(string,keywords):
    keywords_found = []
    for keyword in keywords:
        if string.find(keyword) != -1:
            keywords_found.append(keyword)
    return keywords_found 
```

```python
test_string = "The quick brown fox jumps over the lazy dog"

print(test_string.find("fox"))
print(test_string.find("squirrel"))

find_keywords(test_string,["quick","over","Zwetschkenkuchen"])
```

```python
def join_paragraphs(paragraphs):
    joined_text = ""
    for p in paragraphs:
        joined_text += "\n" + p
    
    return joined_text

text_df = pd.DataFrame(
    {
        "section" : [0,1,1,1,2,2,3,3,3,3],
        "subsection" : [0,0,1,2,0,1,0,1,2,3],
        "text" : [
            "abc","def","ghi","jkl","mno","pqr","stu","vwx","yz","alpha"
        ],
        "more_text" : [
            "abc","def","ghi","jkl","mno","pqr","stu","vwx","yz","alpha"
        ]
    }
)
text_df.head(10)
```
