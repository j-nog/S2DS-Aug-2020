## 1. Import and clean the data

```python
def first_level_pdf_cleaning(doc):
    """
    Cleans the document from unneecessary chars/words, etc.
    """

    doc = str(doc)                               # in case numeric type exsisted 
    doc = doc.lower()                            # concert to lower case letters
    #doc = re.sub(r"(\d|\\W)+", "", doc)          # remove digits and special characters from string
    doc = re.sub(r'[^(\w\s)]','', doc)           # remove punctuation
    doc = re.sub(r"[\t\n]+", "", doc)            # find & replace \t and \n with empty string
    doc = re.sub(r"[^\x00-\x7F]+", " ", doc)     # remove non-ascii chars
    doc = re.sub(r" +", " ", doc)                # remove dublicate spaces
    doc = re.sub(r"(Page)\s\d+\s\w+\s\d+\s(Tiers 2 and 5: guidance for sponsors - version 07\/20)", "", doc) # @Eyzo
    doc = doc.strip()                            # strip leading/trailing spaces
    return doc
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Tier 2 and 5: Guidance for Sponsors")
type(doc)
expression = r"[Tt]ier \d (\, \d)*(and \d)?"
for match in re.finditer(expression, doc.text):
    start, end = match.span()
    span = doc.char_span(start, end)
    
    # This is a Span object or None if match doesn't map to valid token sequence
    if span is not None:
        print("Found match:", span)
        print(re.sub(r"( and )", " and tier ", "Tier 2 and 5: Guidance for Sponsors" ))
        
        
```

```python
df = pd.read_csv("../../pdf_docs_larisa/2020-07-16_Tier-2-5-sponsor-guidance_Jul-2020_v1.0_section.csv", index_col=0)
df.head()
```

```python
df.raw_text = df.raw_text.apply(lambda x:first_level_pdf_cleaning(x))
df.raw_text[0]
```

```python
""" To remove stopwords """
from nltk.corpus import stopwords

stopwords_en = set(stopwords.words('english'))

def remove_stopwords(text):

    text = [word for word in text if word not in stopwords_en]
    
    return text
```

```python
remove_stopwords('Remove all the stop words.'.split())
```

```python
for row in range(len(df)):
    df.at[row, 'raw_text'] = remove_stopwords(df['raw_text'][row].split())
```

```python
""" To lemmatize words """

from nltk.stem.wordnet import WordNetLemmatizer

lem = WordNetLemmatizer()

def wordLemmatizer(text):
    text = [lem.lemmatize(word) for word in text]
    
    return text
```

```python
for row in range(len(df)):
    df['raw_text'][row] = wordLemmatizer(df['raw_text'][row])
```
